```python
student=data.drop(['ses', 'schtyp', 'female'], axis=1)
```

```python
#3 relationship analysis
plt.figure(figsize=(12, 8))
```

```python
correlation = student[['read', 'write', 'math', 'science']].corr()
```

```python
sns.heatmap(correlation, annot=True,  cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Academic Scores')
plt.show()
```

```python
sns.pairplot(student)
```

```python
#4 Scatter Plot
plot1=sns.relplot(x='math', y='science', hue='female', data=data)
plot1.set(title='Math vs Science scores by Gender')

plot2=sns.relplot(x='math', y='science', hue='prog', data=data)
plot2.set(title='Math vs Science Scores by Chosen Programs')

plot3=sns.relplot(x='prog', y='science', hue='female', data=data)
plot3.set(title='Chosen Programs vs Science Scores by Gender')

plot4=sns.relplot(x='prog', y='math', hue='female', data=data)
plot4.set(title='Chosen Programs vs Math Scores by Gender')
```
